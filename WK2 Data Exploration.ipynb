<h1>Mass Shooting Data

<h5>This data set is about gun violence incidents, particularly of massshootings, in the US between the years 2014 to 2016. The data allows me to explore the date, location, killed, and injured of any given mass shooting in those three years. 
   


**Importing Data**

Here I import my data set file,"mass_shootings_all.csv - Sheet1.csv" by using geopandas to import it. I then assign it to the variable "gvd" in order to reference to it more easily and I will know its my gun violence data. 

```python
import geopandas as gpd
```

```python
gvd = gpd.read_file('mass_shootings_all.csv - Sheet1.csv')
```

<h1>Commands

<h5>The following commands I will be doing are specific to geopandas and they are .shape, .info, .head(), .plot(), and .value_counts()


**.head Command-** By adding the command .head to gvd it will output the first five rows of the data set, starting from 0 to 4.   

```python
gvd.head()
```

**.shape Command-**The command .shape. outputs the amount of rows and columns in the whole dataset imported. SO here there are 500 rows and 8 columns. this allows me to know exactly how large my data is and have an idea of what quantity I will work with.

```python
gvd.shape
```

**.info Command-** By using the command .info on gvd you can output essentially a summary of the data set. This allows me to all get an overview of my data and whow what kind of data it is that my data set is comprised and organized into. What is output here is the first and last 5 rows of each column. 

```python
gvd.info
```

**.value_counts Command-** by using the braces I am able to pull only values from the "State" column from the gvd data set but when you add the .value_count() to the "gvd['State']" the output will be how many instances (or value counts) there are for each value. As a result, what is returned is the number of seperate mass shooting incidents have occurred in each state over the duration of the three years. 

```python
gvd['State'].value_counts()
```

<h1>Plotting

Here I am going to create a variable that will comprise of the state value counts titled "stateCount" so as to condense it and make the variable contain a series that can later be converted. I then output "stateCount" to demonstrate/verify that is what happened. 

```python
stateCount = gvd['State'].value_counts()
stateCount
```

I will now convert the series into a dataframe by using ".reset_index()" what is then output is that dataframe of the states and the number of counts 

```python
stateCount= stateCount.reset_index()
stateCount
```
